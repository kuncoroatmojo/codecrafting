```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import metrics
```

```python
dataset = pd.read_excel("titanic3.xls")
dataset.sample(10)
```

```python
dataset.describe()
```

```python
print dataset.isnull().sum()
```

```python
nsurvived = dataset[dataset['survived']==1]
ndead = dataset[dataset['survived']==0]
labels = "survived","dead"
explode = (0.01,0.01)
plt.pie([len(nsurvived),len(ndead)], explode=explode, labels=labels,autopct="%1.1f%%")
plt.show()
```

```python
dataset['age'].fillna(dataset['age'].median(), inplace=True)
replace_embarked = dataset['embarked'].mode()
dataset['embarked'].fillna(str(replace_embarked), inplace=True)
dataset['fare'].fillna(dataset['fare'].median(), inplace=True)
```

```python
print dataset.isnull().sum()
dataset.sample(10)
```

```python
dataset['name_length'] = dataset['name'].apply(len)
dataset['cabin_known'] = np.where(dataset['cabin'].isnull(),0,1)
dataset['boat_known'] = np.where(dataset['boat'].isnull(),0,1)
dataset['body_known'] = np.where(dataset['body'].isnull(),0,1)
dataset['family_size'] = dataset['sibsp'] + dataset['parch'] + 1
dataset['is_alone'] = 0
dataset.loc[dataset['family_size'] == 1,'is_alone'] = 1
dataset['title'] = dataset['name'].str.split(", ", expand=True)[1].str.split(".", expand=True)[0]
dataset['fare_bin'] = pd.qcut(dataset['fare'],3)
dataset['age_bin'] = pd.qcut(dataset['age'].astype(int), 5)
title_min = dataset['title'].value_counts() < 10
dataset['title'] = dataset['title'].apply(lambda x:'Misc' if title_min.loc[x] else x)
print(dataset['title'].value_counts())
dataset['ticket_shared'] = np.where(dataset.groupby('ticket')['name'].transform('count')>1, 1, 0)
```

```python
sns.barplot(x='sex', y='survived', data=dataset)
```

```python
p = sns.pointplot(x='age_bin', y='survived', hue='sex', data=dataset, palette='husl',ci=95.0)
```

```python
print(dataset['age_bin'].value_counts())
```

```python
dataset['sex'] = dataset['sex'].astype('category').cat.codes
dataset = pd.get_dummies(dataset, columns=['pclass', 'embarked', 'cabin_known', 'boat_known', 'body_known', 'title', 'fare_bin', 'age_bin'])
dataset = dataset.drop(labels=['cabin','boat','body','home.dest','ticket','fare','age','name'], axis=1)
```

```python
print dataset.isnull().sum()
dataset.sample(10)
```

```python
colormap = plt.cm.RdBu
plt.figure(figsize=(20,17))
plt.title('Features Correlation', y=1.05, size=15)
sns.heatmap(dataset.astype(float).corr(),linewidths=0.1, vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```
